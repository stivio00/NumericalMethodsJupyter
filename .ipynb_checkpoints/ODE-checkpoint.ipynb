# ODE
Un sistema de ecuaciones diferenciales ordinarias:
$$ \frac{d}{dx}\vec y(x) = \vec f(x,\vec y) $$


Caso escalar:
$$ \frac{d}{dx} y(x) = f(x, y) $$


Usando una aproximacion de Tylor expandida en $i+1$
`Nuevo Valor = antiguo valor + pendiente*paso`
Esto en temrinos matematico lo podemos escribir como
$$ y_{i+1} = y_{i} + \phi h$$

$$ \frac{\mathrm d y}{\mathrm d x} = 
-2 x^3 + 12 x^2 -20 x +8.5 $$

Con solución exacta :
con la condicion inicial: $y(x=0)=1$
$$ y = - 0.5x^4 + 4x^3 - 10x^2 + 8.5 x +1 $$

$$y_i = y(x_i)$$
$$ x_{i+1} = x_i + h $$


```python
import numpy as np
import matplotlib.pylab as plt

def f(x,y):
    return -2*x*x*x + 12*x*x - 20*x + 8.5

def y_real(x,y):
    return -0.5*x**4 + 4*x**3 - 10*x**2 + 8.5*x  + 1
```

```python
h = 0.1 # paso
y_0 = 1
x_0 = 0

y_1 = y_0 + f(x_0, y_0)*h
x_1 = x_0 + h

y_2 = y_1 + f(x_1, y_1)*h
x_2 = x_1 + h
```

```python
xx = np.linspace(0,4) #independiente
#plt.plot(xx, f(xx,0), label="f(x)")
plt.plot(xx, y_real(xx,0), label="Solución exacta")

plt.plot([x_0, x_1, x_2], [y_0, y_1, y_2],"o-", label="Solución aprox. euler")
plt.legend(loc="best")
plt.show()
```

```python
def euler(f, x0, y0,xf, h):
    "f(x,y)"
    l = xf - x0
    qty = int(l/h)
    print("Metodo de euler con h=%f, x0=%f, xf=%f, long=%f, N=%f"%(h,x0,xf,l,qty))
    
    xx = np.linspace(x0,xf, qty+1)
    yy = np.zeros(qty+1)
    
    xx[0] = x0
    yy[0] = y0
    
    i = 1  
    while i <= qty:
        xx[i] = xx[i-1] + h
        yy[i] = yy[i-1] + f(xx[i-1], yy[i-1])*h
        i = i + 1
    
    return np.array([xx,yy])
```

```python
respuesta = euler(f,0.0 ,1.0, 4.0, .5 )
respuesta
```

```python
xx = np.linspace(0,4) #independiente
plt.plot(xx, y_real(xx,0), label="Solución exacta")
plt.plot(respuesta[0], respuesta[1] ,"o-", label="Solución aprox. euler")
plt.legend(loc="best")
plt.show()
```

# Sistemas de ordenes superiores
Sistema oscilatorio harmonico

$$ y'' = -y $$
$$ y = sin(x) $$ 

Los estados:
$$ y_1 = y'$$
$$ y_2 = y$$
Subst:
$$ y_1' = - y_2 $$
$$ y_2' = y_1 $$

$$ \vec y = [y_1  \quad y_2]^t $$
$$ \vec y' = [-y_2 \quad y_1]^t $$

* Sistema de 2 ecuaciones de primer orden:
$$ \vec f(x, \vec y) = [-y_2 \quad y_1]^t $$

Euler en componentes
$$y_{i+1}^0 = y_i^0 + f^0(x,y) h$$
$$y_{i+1}^1 = y_i^1 + f^1(x,y) h$$
Euler en Vectores:
$$\vec y_{i+1} = \vec y_i + \vec f(x,y) h$$

```python
def euler_md(f, x0, y0,xf, h):
    "f(x,y) vector de N elementos"
    l = xf - x0
    qty = int(l/h)
    print("Metodo de euler con h=%f, x0=%f, xf=%f, long=%f, N=%f"%(h,x0,xf,l,qty))
    
    #numero de ecuaciones
    N = len(f(x0,y0)) 
    
    xx = np.linspace(x0,xf, qty+1)
    yy = np.zeros([qty+1, N])
    
    xx[0] = x0
    yy[0] = y0
    
    i = 1  
    while i <= qty:
        xx[i] = xx[i-1] + h
        yy[i] = yy[i-1] + f(xx[i-1], yy[i-1])*hs
        i = i + 1
    
    return [xx,yy]
```

```python
def harmonico_simple(x, y):
    return np.array([-y[1], y[0]])
```

```python
r = euler_md(harmonico_simple, 0.0, [0.0, 1.0] , 5.0, .01)
plt.plot(r[0], r[1][:,0], label="y'")
plt.plot(r[0], r[1][:,1], label="y")
r
```

## Sistema masa, resorte y amortiguador
$$ \vec y = [y^1 \quad y^0]^T $$

$$ \vec f(x, \vec y ) = \begin{bmatrix}-c/m \cdot y^1 - k/m \cdot y^0  \\ y^1 \end{bmatrix} $$


```python
def mra(x,y, args=[1.0, 0.0, 1.0]):
    #args = (m,c,k)
    m = args[0]
    c = args[1]
    k = args[2]
    return np.array([-c/m*y[0] -k/m*y[1], y[0]])

r = euler_md(mra,0.0, [0.0,1.0], 10.0, .1)
t = r[0]
x = r[1][:,1]
v = r[1][:,0]
plt.plot(t,x, label="pos")
plt.plot(t,v, label="vel")
plt.legend(loc="best")
```
