Calculo simbolico con Python

## Introduccion


```python
%matplotlib inline 
import sympy as sp
sp.init_printing()

x,y,z = sp.symbols("x y z")
```

```python
f = x**2 -4
f2 = sp.exp(x) - x**2
```

```python
[s] = sp.solve(f2, x)
s.n()
```

```python
sp.plot(f)
sp.plot(f2)
sp.plot(sp.LambertW)
```

```python
v = sp.LambertW(2)
```

```python
sp.LambertW(-.5).n()
```

```python
f2.diff(x,3)
```
